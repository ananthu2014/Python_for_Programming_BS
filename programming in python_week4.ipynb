```python
class a:
    def __init__(self,a):
        self.a = a
    
    def __len__(self):
        return len(self.a)
    
    def len(self):
        return len(self.a)
    
    def __getitem__(self,index):
        return self.a[index]
```

```python
class MyClass:
    def __init__(self, value):
        self.a = value  # Assign the passed value to the instance attribute

    def __add__(self, other):
        if isinstance(other, MyClass):  # Check if 'other' is an instance of MyClass
            return self.a + other.a
        else:
            raise ValueError("The 'other' object is not an instance of MyClass")
# Create two instances of MyClass
instance_1 = MyClass(10)
instance_2 = MyClass(20)

# Perform the addition
result = instance_1 + instance_2  # This triggers MyClass.__add__
print(f"Result of addition: {result}")  # Expected output: 30
```

```python
n = int(input())
lis=[]
for i in range(n):
    coord = list(map(int,input().split(',')))
    lis.append(coord)
    
point = list(map(int,input().split(',')))
dict1={}
for i in range(len(lis)):
    key = tuple(lis[i])
    dist = ((point[0] - lis[i][0])**2 + (point[1] - lis[i][1])**2)**0.5
    dict1[key] = dist

minimum = min(dict1.values())
for i in dict1.items():
    if i[1] == minimum:
        print(*i[0],sep=',')
```
